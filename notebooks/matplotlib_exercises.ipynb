```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# basic line plot

# import data
infile = '../../kuroshio_productivity/data/jma_nitrate.csv'
data = pd.read_csv(infile, sep=',')

data.columns
```

```python
plt.plot(data.lat, data.nitrate, '.')
plt.show()

subset = data[data.nitrate> 0]
plt.plot(subset.lat, subset.nitrate, '.')
plt.xlabel('latitude')
plt.ylabel('nitrate')
plt.show()
```

```python
fig = plt.figure(figsize=(5, 4), dpi = 300)

font = {'family': 'sansserif',
        'color':  'purple',
        'weight': 'normal',
        'size': 16,
        }

plt.plot(subset.lat, subset.nitrate, '.k')
plt.xlabel('latitude',fontdict=font)
plt.ylabel('nitrate',fontdict=font)

plt.title('random data', fontsize = 20, color = 'red', fontname = 'Times')

plt.show()

fig.savefig('example.png', dpi = 300)
```

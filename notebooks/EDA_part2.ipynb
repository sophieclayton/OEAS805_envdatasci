```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
infile = '../data/city_temperature.csv'
```

```python
data = pd.read_csv(infile, sep = ',', low_memory = False)
```

```python
data.columns
```

```python
data['Country'].unique()
```

```python
cr = data[data['Country'] == 'Costa Rica']
```

```python
cr.info()
```

```python
crm = cr.groupby('Month')['AvgTemperature'].mean()
```

```python
sns.lineplot(x = 'Year', y = 'AvgTemperature', data = cr)
```

```python
sns.barplot(data = cr, x = 'City', y= 'AvgTemperature')
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# import the data 
data = pd.read_csv('jma_nitrate.csv')
data.columns

data = data[data['nitrate']>0]
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

y = data.nitrate
X = data[['density', 'depth']]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

print('r2 =', r2_score(y_train, regr.predict(X_train)))
print('r2 =', r2_score(y_test, y_pred))
print('MSE =',mean_squared_error(y_test, y_pred))
```

```python
plt.plot(X_test['density'], y_pred, 'xk', alpha = 0.8)
plt.plot(X_train['density'], y_train, 'or', zorder = 0)
plt.show()

plt.plot(X_test['depth'], y_pred, 'xk', alpha = 0.8)
plt.plot(X_train['depth'], y_train, 'or', zorder = 0)
plt.show()
```
